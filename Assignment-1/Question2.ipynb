```python
import custom_func
from collections import Counter
```

```python
with open('abc.txt','r',encoding='utf-8') as f:
    corpus=f.read()
```

```python
characters=custom_func.char_list(corpus[:1000])
characters
```

```python
char_frq=Counter(characters)
char_frq=custom_func.sort_dict(char_frq)
char_frq
```

```python
#Top-20 Character:
dict(list(char_frq.items())[:20])
```

```python
syllables=custom_func.syll_list(corpus[:1000])
syllables
```

```python
syll_frq=Counter(syllables)
syll_frq=custom_func.sort_dict(syll_frq)
syll_frq
```

```python
#Top 20 Syllables
dict(list(syll_frq.items())[:20])
```
