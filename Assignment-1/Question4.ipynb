```python
import sentencepiece as st
from collections import Counter
import custom_func
```

```python
filepath='abc.txt'

with open(filepath,'r',encoding='utf-8') as f:
    corpus=f.read()
```

# **Part 1: Unigram**

```python
st.SentencePieceTrainer.train(input=filepath, model_prefix='unigram_model_1K', vocab_size=1000, model_type='unigram')
```

```python
st_uni = st.SentencePieceProcessor(model_file='unigram_model_1K.model')
```

```python
tokens = st_uni.encode_as_pieces(corpus)
```

```python
#Preprocessing of tokens
for i in range(len(tokens)):
    if tokens[i].startswith('▁'):
        tokens[i] = tokens[i][1:]
```

```python
len(tokens)
```

```python
#Unigram Frequency
uni_frq_tokens=custom_func.uni_dict(tokens)
uni_frq_tokens
```

```python
#Bigram Frequency
bi_frq_tokens = custom_func.bi_dict(tokens)
bi_frq_tokens
```

```python
#syllables
syllables=custom_func.syll_list(tokens)
syllables
```

```python
#Unigram Frequency of Syllables
uni_frq_syllables=custom_func.uni_dict(syllables)
uni_frq_syllables
```

```python
#Bigram Frequency of Syllables
bi_frq_syllables=custom_func.bi_dict(syllables)
bi_frq_syllables
```

```python
#character
characters=custom_func.char_list(tokens)
characters
```

```python
#Unigram Frequency of Characters
uni_frq_char=custom_func.uni_dict(characters)
uni_frq_char
```

```python
#Bigram Frequency of Characters
bi_frq_char=custom_func.bi_dict(characters)
bi_frq_char
```
