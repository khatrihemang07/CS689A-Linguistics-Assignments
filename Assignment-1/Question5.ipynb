```python
import sentencepiece as st
from collections import Counter
import custom_func
```

```python
def bi_list(tokens):
    #Bigram Frequency
    bigrams = [f"{tokens[i]} {tokens[i+1]}" for i in range(len(tokens)-1)]
    return bigrams
```

```python
with open('truth.txt','r',encoding='utf-8') as f:
    truth=f.read()
truth
```

```python
truth=truth.split(',')
ab=[]
for a in truth:
    ab.append(a.strip())
truth=ab
```

```python
with open('q1.txt','r',encoding='utf-8') as f:
    questions=f.read()   
questions
```

# **Unigram**

```python
st_uni = st.SentencePieceProcessor(model_file='unigram_model_1K.model')
```

```python
tokens = st_uni.encode_as_pieces(questions)
```

```python
#Preprocessing of tokens
for i in range(len(tokens)):
    if tokens[i].startswith('▁'):
        tokens[i] = tokens[i][1:]
```

```python
len(tokens)
```

```python
#syllables
syllables=custom_func.syll_list(tokens)
syllables
```

```python
#character
characters=custom_func.char_list(tokens)
characters
```

```python
#Precision,recall and F-score for Unigram of tokens
common=set(truth)& set(tokens)
precision=len(common)/len(tokens)
recall = len(common)/len(truth)
if precision + recall != 0:
    f_score = 2 * (precision * recall) / (precision + recall)
else:
    f_score = 0

print("Precision: ",precision)
print("Recall: ",recall)
print("F-score: ",f_score)
```

```python
len(common)
```

```python
#Precision for Bigram of tokens
bi_tokens=bi_list(tokens)
common=set(truth)& set(bi_tokens)
precision=len(common)/len(bi_tokens)
recall = len(common)/len(truth)
if precision + recall != 0:
    f_score = 2 * (precision * recall) / (precision + recall)
else:
    f_score = 0

print("Precision: ",precision)
print("Recall: ",recall)
print("F-score: ",f_score)
```
