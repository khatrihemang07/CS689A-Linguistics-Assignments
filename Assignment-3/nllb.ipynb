```python
from transformers import AutoModelForSeq2SeqLM,AutoTokenizer
tokenizer=AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-600M")
model=AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")
```

# Loading English Data

```python
with open('/kaggle/input/dataset/test.en','r',encoding='utf-8') as f:
    english_sentences=f.readlines()

english_sentences[:10]
```

# Loading Hindi Data

```python
with open('/kaggle/input/dataset/test.hi','r',encoding='utf-8') as f:
    hindi_sentences=f.readlines()

hindi_sentences[:10]
```

# Loading Gujarati Data

```python
with open('/kaggle/input/dataset/test.gu','r',encoding='utf-8') as f:
    gujarati_sentences=f.readlines()

gujarati_sentences[:10]
```

# Random Sampling 1000 examples

```python
from random import sample,seed
seed(1748)
sample_indices=sample(range(len(english_sentences)),1000)
sample_indices[:10]
```

```python
english_sentences=[english_sentences[i] for i in sample_indices]
hindi_sentences=[hindi_sentences[i] for i in sample_indices]
gujarati_sentences=[gujarati_sentences[i] for i in sample_indices]
```

```python
english_sentences[:10]
```

```python
hindi_sentences[:10]
```

```python
gujarati_sentences[:10]
```

# Translating English -> Hindi

```python
translated_english_hindi=[]
```

```python
for i in range(0,25):
    batch=tokenizer(english_sentences[i*40:(i*40)+40],return_tensors='pt',padding='longest')
    outputs=model.generate(**batch,forced_bos_token_id=tokenizer.lang_code_to_id['hin_Deva'],max_length=30)
    translated_english_hindi.extend(tokenizer.batch_decode(outputs,skip_special_tokens=True))
```

```python
translated_english_hindi[:10]
```

```python
with open("/kaggle/working/translated_english_hindi.txt",'w',encoding='utf-8') as f:
    for s in translated_english_hindi:
        f.write(s+'\n')
```
