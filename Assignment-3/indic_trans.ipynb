# Importing IndicTrans

```python
import torch
from transformers import AutoModelForSeq2SeqLM
from IndicTransTokenizer import IndicProcessor, IndicTransTokenizer
```

# Creating Tokenizer and Model for English -> Indic

```python
tokenizer_english=IndicTransTokenizer(direction='en-indic')
model_english=AutoModelForSeq2SeqLM.from_pretrained("ai4bharat/indictrans2-en-indic-dist-200M",trust_remote_code=True)
```

# Creating Tokenizer and Model for Indic -> English

```python
tokenizer_indic=IndicTransTokenizer(direction='indic-en')
model_indic=AutoModelForSeq2SeqLM.from_pretrained("ai4bharat/indictrans2-indic-en-dist-200M",trust_remote_code=True)
```

# Creating Tokenizer and Model for Indic -> Indic

```python
tokenizer_indic_indic=IndicTransTokenizer(direction='indic-indic')
model_indic_indic=AutoModelForSeq2SeqLM.from_pretrained("ai4bharat/indictrans2-indic-indic-dist-320M",trust_remote_code=True)
```

```python
processor=IndicProcessor(inference=True)
```

# Loading English Data

```python
with open('/kaggle/input/dataset/test.en','r',encoding='utf-8') as f:
    english_sentences=f.readlines()

english_sentences[:10]
```

# Translating English -> Hindi

```python
translated_english_hindi=[]
```

```python
for i in range(0,25):
    batch=processor.preprocess_batch(english_sentences[i*40:(i*40)+40],src_lang='eng_Latn',tgt_lang='hin_Deva')
    batch=tokenizer_english(batch,src=True,return_tensors='pt')
    with torch.inference_mode():
        outputs=model_english.generate(**batch,num_beams=5,num_return_sequences=1,max_length=256)
    outputs=tokenizer_english.batch_decode(outputs,src=False)
    translated_english_hindi.extend(processor.postprocess_batch(outputs,lang='hin_Deva'))
```

```python
(translated_english_hindi)[:10]
```

```python
len(translated_english_hindi)
```

```python
with open("/kaggle/working/translated_english_hindi.txt",'w',encoding='utf-8') as f:
    for s in translated_english_hindi:
        f.write(s+'\n')
```
